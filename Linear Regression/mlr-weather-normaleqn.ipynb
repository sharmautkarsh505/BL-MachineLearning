```python
X=data[['Temperature (C)','Humidity','Wind Speed (km/h)','Visibility (km)','Pressure (millibars)']]
y=data['Apparent Temperature (C)']
```

```python
ones=np.ones((96453,))
X.insert(loc=0,column='ones',value=ones)
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
y_test.shape
```

```python
X_train
```

```python
#basic formula for obaining parameters
#inverse of(X_train_T*X_train)*(X_train_T*y_train) where X_train_T is the transpose of X_train
X_train_T=X_train.transpose()
product=X_train_T.dot(X_train)
product.to_numpy()
inverse_product=np.linalg.inv(product)

product1=X_train_T.dot(y_train)

final_product=inverse_product.dot(product1)
np.set_printoptions(suppress=True)
print(final_product)
```

```python
y_pred_testing=(-2.50045233)+(1.12531219)*X_test.iloc[:,0]+(1.03283023)*X_test.iloc[:,1]+(-0.0950324)*X_test.iloc[:,2]+(-0.00089798)*X_test.iloc[:,3]+(0.0002069)*X_test.iloc[:,4]

y_testing_mean=y_test.mean()

tss=sum((y_test-y_testing_mean)**2)
rss=sum((y_pred_testing-y_test)**2)

r2score=1-(rss/tss)
print(r2score)
```

```python
df=pd.DataFrame(data=y_pred_testing,columns=['predicted y'],index=list(range(28936)))
df.insert(loc=0,column='y test',value=y_test)

print(df.to_string())
```
